```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/marketing_data (1).xlsx")
```

```python
df
```

```python
df.info()
```

```python
df["Age"]= 2025-df["Year_Birth"]
```

```python
df["Age_Group"]= pd.cut(df["Age"],
                        bins =[0,18,40,60,123],
                        labels=["Child","Adult","Senior","Old Age"])
```

```python
df["Total_Spending"]=(df['MntWines']+df['MntFruits']+df['MntMeatProducts']+
                      df['MntFishProducts']+ df['MntSweetProducts']+df['MntGoldProds'])
```

```python
df["Total_Reg_Purchase"]= (df['NumWebPurchases']+df['NumCatalogPurchases']+
                           df['NumStorePurchases'])
```

```python
df["Purchases_Type"]= np.where(df['NumDealsPurchases']>0,"Deal","Regular")
```
