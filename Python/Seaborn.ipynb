```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head(1)
```

```python
# histogram / distribution of coninious numerical feature
```

```python
sns.histplot(data = titanic,x = 'age' ,bins=30, color ='skyblue')
plt.title("Distribution of age ")
```

```python
# Count Plot  - categorical count
```

```python
titanic['sex'].value_counts().index
```

```python
titanic['sex'].value_counts().values

```

```python
sns.countplot(data = titanic,x ='sex',hue='survived',palette= 'Set1')
```

```python
#bar plot
```

```python
sns.barplot(data = titanic,x='pclass',y='fare',ci=False)
```

```python
#Box plot
```

```python
sns.boxplot(data=titanic,x='pclass',y='age',hue='sex',palette='viridis')
```

```python
sns.violinplot(data=titanic,x='pclass',y='age',hue='sex',palette='viridis')
```

```python
#Scatter plot
```

```python
sns.scatterplot(data = titanic, x="age",y="fare",hue="sex" ,palette='hls')
```

```python
#heatmap - always for correlation of numerical
```

```python
corr = titanic[['age','fare','pclass','sibsp','parch']].corr()
```

```python
sns.heatmap(corr, annot=True,cmap='coolwarm')
```

```python
sns.kdeplot(data = titanic , x='age')
```

```python
sns.pairplot(titanic)
```

```python
sns.swarmplot(data=titanic,x='pclass',y='age',hue='sex')
```
