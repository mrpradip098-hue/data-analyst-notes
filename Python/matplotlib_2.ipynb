```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


```

```python
# BoxPlot - outlier finding
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head(1)
```

```python
plt.figure(figsize=(5,4))
titanic.boxplot(column='fare', by ='pclass',grid=False)
plt.savefig("bc.jpeg")
```

```python
#scatter plot - Num Vs Num -
```

```python
plt.figure(figsize=(6,4))
plt.scatter(titanic['age'],titanic['fare'],c="red")
plt.title("scatter plot of age vs fare")
plt.xlabel("Age of passsenger")
plt.ylabel("Fare of passenger")
plt.show()
```

# Charts using SNS seaborn library

syntax
1 - set the figure size using plt

2 - sns. graph type name (data= df,x = column 1 , y = column 2, hue= column 3)

3 - plt.title()

4 - plt.xlabel()

5 - plt.ylabel()

6 - plt.legend()

7 - plt.show()


```python
plt.figure(figsize=(6,4))
sns.scatterplot(data = titanic,x='age',y='fare' , hue = 'sex')
plt.title("scatter plot")
```

```python
#lineplot - correlation analysis
```

```python
sns.lineplot(data = titanic, x = 'age', y = 'fare',  hue='sex')
plt.title("line plot of age vs fare")
```

```python
sns.boxplot(data=titanic, x='pclass',y='fare',hue='sex')
```
